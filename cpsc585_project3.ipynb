#### **Experiment 5**

Implement a multiclass perceptron as described in Section 2.3.1 of the textbook to recognize the letters A-Z.
Begin with weights initialized to random values and a learning rate of ùõº = 0.01. Predict the letter by choosing the class with the largest WÃÖr‚àôXÃÖi value.
If the prediction is incorrect, train the multiclass perceptron by applying the update rule of Equation 2.24 in the textbook. Continue training until all items in the training set are correctly classified.

```python
input_dim = (len(TRAINING_SET[0][0])) + 1
perceptron_num = len(string.ascii_uppercase)

weights_matrix = rng.uniform(-1, +1, (perceptron_num, input_dim))
weights_matrix_2 = rng.uniform(-1, +1, (perceptron_num, input_dim))

learn_rate = 0.01
epoch_n = 1000

correct = 0
incorrect = 0

for epoch in range(epoch_n):
  for i in range(len(train_features)):
    y_i = np.where(train_labels[i])
   
    current_activation = []
    arg_max = 0
    predicted_y = 0

    # Multi-Class Decision Rule
    for label in train_labels:
        current_activation = np.dot(weights_matrix[np.where(label)], train_features[i])
        if current_activation > arg_max:
          arg_max = current_activation
          predicted_y = label
    
    # Update Rule
    if not (np.where(train_labels[i]) == np.where(predicted_y)):
      weights_matrix[y_i] += (learn_rate*train_features[i])
      weights_matrix[np.where(predicted_y)] -= (learn_rate*train_features[i])
      incorrect += 1
    elif (y_i == np.where(predicted_y)):
      correct += 1
    print('------------------------------------')
    print('Epoch ' + str(epoch + 1))
    print('------------------------------------')
    print('Accuracy: ' + str(correct/(correct + incorrect)))
```
