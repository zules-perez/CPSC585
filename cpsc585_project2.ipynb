### Experiment 2

In order to use the images in TRAINING_SET, TEST_SET, and MESSAGE, convert them into two-dimensional NumPy arrays of feature vectors.

In order to evaluate the output of your code, you may wish to consider implementing a Python function show(image) to print() a letter image row-by-row using a hash mark ('#') for 1 and a space (' ') for 0 and testing this function on some of the images in TRAINING_SET.


```python
import numpy as np
    
train_features = np.array([i[0] for i in TRAINING_SET])
test_features = np.array([i[0] for i in TEST_SET])

message = np.array(MESSAGE)

def show(image):
    column = 0
    for character in image:
        column += 1
        if character == 1:
            print("#", end = '')
        else:
            print(' ', end = '')
        if column == 5: # Each font matrix has 7 rows and 5 columns
            column = 0 # Reset column to 0 to process next row
            print('')

for i in range(len(train_features)):
    show(train_features[i])
    print('\n')
```

### Experiment 3

In order to use the character labels in TRAINING_SET and TEST_SET, convert them into an integer class vector using ord(), then into one-hot encoded categorical features.

```python
print(train_features.shape)
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import to_categorical

train_ = np.array([(ord(i[1])-65) for i in TRAINING_SET])
test_ = np.array([(ord(i[1])-65) for i in TEST_SET])

train_labels = np.array(to_categorical(train_))

test_labels = np.array(to_categorical(test_))
```

### Experiment 4

Create a Sequential Keras model with a Dense hidden layer and an output layer with softmax activation.

```python

model = keras.Sequential()
model.add(Dense(512, activation = 'relu', input_shape = (35, ), name = 'input_layer1'))
model.add(Dropout(0.2))
model.add(Dense(256, activation = 'relu', name = 'hidden_layer2'))
model.add(Dropout(0.1))
model.add(Dense(26, activation = 'softmax', name = 'output_layer3'))
model.summary()
```

### Experiment 5

compile and fit the model to the training set. Train the model until the accuracy is as high as possible. You may wish to use an EarlyStopping callback.

```python
from keras import callbacks
from keras.optimizers import Adam
earlystopping = callbacks.EarlyStopping(
    monitor = 'loss',  
    patience = 5,
    mode = 'min'
)

model.compile(
    optimizer = 'rmsprop',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)


model.fit(
    train_features,
    train_labels,
    epochs = 500,
    batch_size = 26,
    callbacks = [earlystopping],
    
)
```

### Experiment 6

evaluate the model on TEST_SET.

#### What accuracy do you obtain? 
Accuracy : 0.9615

#### If the accuracy is less than 100%, which test images are misclassified?

The image with the letter W has been misclassified as a U, V, or Y.

```python
model.evaluate(
    test_features,
    test_labels
)

#model.save_weights('model.h5')
#model.load_weights('model.h5')

predict_test = model.predict(test_features[:])

letters = (np.argmax(predict_test, axis = 1) + 65)
print('\n')
for i in range(len(predict_test)):
    print(chr(letters[i]))
    
```
